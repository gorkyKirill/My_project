```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import LabelEncoder
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import TensorBoard


data_popolnenie = [
    (["+", "3", "200", "Баланс:", "25", "535,75", "₽", "МИР", "••", "0919"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "3", "650", "Баланс:", "29", "185,75", "₽", "МИР", "••", "0919"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "2", "823", "Баланс:", "12", "008,75", "₽", "МИР", "••", "0919"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "3", "000.00", "₽.", "Теперь", "на", "карте", "86", "270.48", "₽"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "3", "466.37", "₽.", "Теперь", "на", "карте", "89", "736.85", "₽"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "3", "320", "Баланс:", "15", "328,75", "₽", "МИР", "••", "0919"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "6", "738,46", "Баланс:", "22", "067,21", "₽", "МИР", "••", "0919"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "3", "164,70", "Баланс:", "25", "231,91", "₽", "МИР", "••", "0919"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "2", "550", "Баланс:", "20", "781,91", "₽", "МИР", "••", "0919"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "3", "300", "Баланс:", "6", "081,91", "₽", "МИР", "••", "0919"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "18", "000", "Баланс:", "24", "081,91", "₽", "МИР", "••", "0919"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "2", "850", "Баланс:", "26", "931,91", "₽", "МИР", "••", "0919"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "3", "389", "Баланс:", "10", "320,91", "₽", "МИР", "••", "0919"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "2", "989,27", "Баланс:", "13", "310,18", "₽", "МИР", "••", "0919"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "6", "175,50", "Баланс:", "19", "485,68", "₽", "МИР", "••", "0919"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "9", "000", "Баланс:", "28", "485,68", "₽", "МИР", "••", "0919"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-0997", "21:43", "зачисление", "3450р", "TINKOFF", "Баланс:", "31", "271.1р"], ["O", "O", "B-DEPOSIT", "B-AMOUNT", "O", "O", "O", "O"]),
    (["MIR-0997", "21:45", "Перевод", "7200р", "от", "Михаил", "Л.", "Баланс:", "38", "471.1р"], ["O", "O", "B-DEPOSIT", "B-AMOUNT", "O", "O", "O", "O", "O", "O"]),
    (["MIR-0997", "21:45", "Перевод", "2700р", "от", "Николай", "П.", "Баланс:", "41", "171.1р"], ["O", "O", "B-DEPOSIT", "B-AMOUNT", "O", "O", "O", "O", "O", "O"]),
    (["MIR-0937", "09:10", "зачисление", "2340р", "QIWI", "Баланс:", "43", "511.1р"], ["O", "O", "B-DEPOSIT", "B-AMOUNT", "O", "O", "O", "O"]),
    (["MIR-6894", "11:10", "зачисление", "2340р", "от", "Тинькофф", "Баланс:", "11", "011.12р"], ["O", "O", "B-DEPOSIT", "B-AMOUNT", "O", "O", "O", "O", "O", "O"]),
    (["MIR-6991", "21:30", "перевод", "2340р", "Вася", "Пупкин", "Баланс:", "12", "111.12р"], ["O", "O", "B-DEPOSIT", "B-AMOUNT", "O", "O", "O", "O", "O", "O"]),
        (["+", "3", "200", "Баланс:", "25", "535,75", "₽", "МИР", "••", "0919"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "3", "650", "Баланс:", "29", "185,75", "₽", "МИР", "••", "0919"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "2", "823", "Баланс:", "12", "008,75", "₽", "МИР", "••", "0919"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "3", "000.00", "₽.", "Теперь", "на", "карте", "86", "270.48", "₽"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "3", "466.37", "₽.", "Теперь", "на", "карте", "89", "736.85", "₽"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "3", "320", "Баланс:", "15", "328,75", "₽", "МИР", "••", "0919"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "6", "738,46", "Баланс:", "22", "067,21", "₽", "МИР", "••", "0919"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "3", "164,70", "Баланс:", "25", "231,91", "₽", "МИР", "••", "0919"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "2", "550", "Баланс:", "20", "781,91", "₽", "МИР", "••", "0919"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "3", "300", "Баланс:", "6", "081,91", "₽", "МИР", "••", "0919"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "18", "000", "Баланс:", "24", "081,91", "₽", "МИР", "••", "0919"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "2", "850", "Баланс:", "26", "931,91", "₽", "МИР", "••", "0919"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "3", "389", "Баланс:", "10", "320,91", "₽", "МИР", "••", "0919"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "2", "989,27", "Баланс:", "13", "310,18", "₽", "МИР", "••", "0919"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "6", "175,50", "Баланс:", "19", "485,68", "₽", "МИР", "••", "0919"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "9", "000", "Баланс:", "28", "485,68", "₽", "МИР", "••", "0919"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-0997", "21:43", "зачисление", "3450р", "TINKOFF", "Баланс:", "31", "271.1р"], ["O", "O", "B-DEPOSIT", "B-AMOUNT", "O", "O", "O", "O"]),
    (["MIR-0997", "21:45", "Перевод", "7200р", "от", "Михаил", "Л.", "Баланс:", "38", "471.1р"], ["O", "O", "B-DEPOSIT", "B-AMOUNT", "O", "O", "O", "O", "O", "O"]),
    (["MIR-0997", "21:45", "Перевод", "2700р", "от", "Николай", "П.", "Баланс:", "41", "171.1р"], ["O", "O", "B-DEPOSIT", "B-AMOUNT", "O", "O", "O", "O", "O", "O"]),
    (["MIR-0937", "09:10", "зачисление", "2340р", "QIWI", "Баланс:", "43", "511.1р"], ["O", "O", "B-DEPOSIT", "B-AMOUNT", "O", "O", "O", "O"]),
    (["MIR-6894", "11:10", "зачисление", "2340р", "от", "Тинькофф", "Баланс:", "11", "011.12р"], ["O", "O", "B-DEPOSIT", "B-AMOUNT", "O", "O", "O", "O", "O", "O"]),
    (["MIR-6991", "21:30", "перевод", "2340р", "Вася", "Пупкин", "Баланс:", "12", "111.12р"], ["O", "O", "B-DEPOSIT", "B-AMOUNT", "O", "O", "O", "O", "O", "O"]),
    (["+", "3", "700", "Баланс:", "15", "535,75", "₽", "МИР", "••", "0920"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "3", "900", "Баланс:", "19", "785,75", "₽", "МИР", "••", "0920"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "2", "923", "Баланс:", "13", "108,75", "₽", "МИР", "••", "0920"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "3", "100.00", "₽.", "Теперь", "на", "карте", "87", "370.48", "₽"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "3", "366.37", "₽.", "Теперь", "на", "карте", "90", "836.85", "₽"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "3", "420", "Баланс:", "16", "428,75", "₽", "МИР", "••", "0920"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "6", "838,46", "Баланс:", "23", "167,21", "₽", "МИР", "••", "0920"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "3", "264,70", "Баланс:", "26", "331,91", "₽", "МИР", "••", "0920"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "2", "650", "Баланс:", "21", "881,91", "₽", "МИР", "••", "0920"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "3", "400", "Баланс:", "7", "181,91", "₽", "МИР", "••", "0920"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "18", "100", "Баланс:", "25", "181,91", "₽", "МИР", "••", "0920"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "2", "950", "Баланс:", "27", "031,91", "₽", "МИР", "••", "0920"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "3", "489", "Баланс:", "11", "420,91", "₽", "МИР", "••", "0920"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "2", "089,27", "Баланс:", "14", "410,18", "₽", "МИР", "••", "0920"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "6", "275,50", "Баланс:", "20", "585,68", "₽", "МИР", "••", "0920"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"]),
    (["+", "9", "100", "Баланс:", "29", "585,68", "₽", "МИР", "••", "0920"], ["B-DEPOSIT", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O"])
]

data_spisanie = [
    (["MIR-3893", "00:26", "покупка", "5000р", "TINKOFF", "Баланс:", "13", "661.27р"], ["O", "O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-3993", "00:28", "покупка", "2340р", "TINKOFF", "Баланс:", "11", "321.2р"], ["O", "O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-3993", "21:43", "покупка", "3450р", "TINKOFF", "Баланс:", "5", "271.1р"], ["O", "O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-3894", "01:26", "покупка", "6000р", "TINKOFF", "Баланс:", "14", "661.27р"], ["O", "O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-3994", "02:28", "покупка", "3340р", "TINKOFF", "Баланс:", "12", "321.2р"], ["O", "O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-3994", "22:43", "покупка", "4450р", "TINKOFF", "Баланс:", "6", "271.1р"], ["O", "O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-3995", "23:43", "покупка", "5450р", "TINKOFF", "Баланс:", "7", "271.1р"], ["O", "O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-3995", "24:43", "покупка", "1340р", "TINKOFF", "Баланс:", "8", "271.1р"], ["O", "O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-3996", "22:45", "покупка", "7200р", "TINKOFF", "Баланс:", "9", "271.1р"], ["O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-3996", "22:46", "покупка", "2700р", "TINKOFF", "Баланс:", "10", "271.1р"], ["O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-3996", "22:47", "покупка", "2200р", "TINKOFF", "Баланс:", "11", "271.1р"], ["O", "O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-3997", "22:48", "покупка", "2300р", "TINKOFF", "Баланс:", "12", "271.1р"], ["O", "O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-3898", "00:27", "покупка", "5100р", "TINKOFF", "Баланс:", "14", "761.27р"], ["O", "O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-3998", "00:29", "покупка", "2440р", "TINKOFF", "Баланс:", "12", "421.2р"], ["O", "O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-3998", "21:44", "покупка", "3550р", "TINKOFF", "Баланс:", "6", "371.1р"], ["O", "O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-3899", "01:27", "покупка", "6100р", "TINKOFF", "Баланс:", "15", "761.27р"], ["O", "O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-3999", "02:29", "покупка", "3440р", "TINKOFF", "Баланс:", "13", "421.2р"], ["O", "O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-3999", "22:44", "покупка", "4550р", "TINKOFF", "Баланс:", "7", "371.1р"], ["O", "O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-3999", "23:44", "покупка", "5550р", "TINKOFF", "Баланс:", "8", "371.1р"], ["O", "O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-3999", "24:44", "покупка", "1440р", "TINKOFF", "Баланс:", "9", "371.1р"], ["O", "O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-3000", "22:46", "покупка", "7300р", "TINKOFF", "Баланс:", "10", "371.1р"], ["O", "O", "O", "O", "O", "O"]),
    (["MIR-3000", "22:47", "покупка", "2800р", "TINKOFF", "Баланс:", "11", "371.1р"], ["O", "O", "O", "O", "O", "O"]),
    (["MIR-3000", "22:48", "покупка", "2300р", "TINKOFF", "Баланс:", "12", "371.1р"], ["O", "O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-3001", "22:49", "покупка", "2400р", "TINKOFF", "Баланс:", "13", "371.1р"], ["O", "O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-3991", "00:24", "покупка", "5000р", "TINKOFF", "Баланс:", "13", "661.27р"], ["O", "O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-3992", "00:25", "покупка", "2340р", "TINKOFF", "Баланс:", "11", "321.2р"], ["O", "O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-3992", "21:42", "покупка", "3450р", "TINKOFF", "Баланс:", "5", "271.1р"], ["O", "O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-3893", "01:25", "покупка", "6000р", "TINKOFF", "Баланс:", "14", "661.27р"], ["O", "O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-3993", "02:27", "покупка", "3340р", "TINKOFF", "Баланс:", "12", "321.2р"], ["O", "O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-3993", "22:42", "покупка", "4450р", "TINKOFF", "Баланс:", "6", "271.1р"], ["O", "O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-3994", "23:42", "покупка", "5450р", "TINKOFF", "Баланс:", "7", "271.1р"], ["O", "O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-3994", "24:42", "покупка", "1340р", "TINKOFF", "Баланс:", "8", "271.1р"], ["O", "O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-3995", "22:44", "покупка", "7200р", "TINKOFF", "Баланс:", "9", "271.1р"], ["O", "O", "O", "O", "O", "O"]),
    (["MIR-3995", "22:45", "покупка", "2700р", "TINKOFF", "Баланс:", "10", "271.1р"], ["O", "O", "O", "O", "O", "O"]),
    (["MIR-3995", "22:46", "покупка", "2200р", "TINKOFF", "Баланс:", "11", "271.1р"], ["O", "O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-3996", "22:47", "покупка", "2300р", "TINKOFF", "Баланс:", "12", "271.1р"], ["O", "O", "O", "O", "O", "O", "O", "O"]),
    (["− 7 000.00 ₽ по номеру телефона +79916015674, Нодирбек Одилжон угли А. через СБП, теперь на счете 23 643.65 ₽"], ["O", "B-WITHDRAW", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O"]),
    (["− 3 500.50 ₽ списано с карты **** 5678, Сидоров Петр Петрович, теперь на счете 25 143.15 ₽"], ["O", "B-WITHDRAW", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O"]),
    (["− 1 200.00 ₽ по номеру телефона +79987654321, Алибек Алибекович, через СБП, теперь на счете 33 943.15 ₽"], ["O", "B-WITHDRAW", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O"]),
    (["− 2 300.75 ₽ списано с карты **** 3456, Иванова Елена Сергеевна, теперь на счете 31 642.40 ₽"], ["O", "B-WITHDRAW", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O"]),
    (["− 8 000.00 ₽ по номеру телефона +79912345678, Смирнов Алексей Александрович, через СБП, теперь на счете 23 642.40 ₽"], ["O", "B-WITHDRAW", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O"]),
    (["− 6 750.25 ₽ списано с карты **** 7890, Ковалев Дмитрий Иванович, теперь на счете 16 892.15 ₽"], ["O", "B-WITHDRAW", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O"]),
    (["− 4 100.00 ₽ по номеру телефона +79911223344, Андреев Павел Петрович, через СБП, теперь на счете 12 792.15 ₽"], ["O", "B-WITHDRAW", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O"]),
    (["− 9 500.00 ₽ списано с карты **** 6543, Федоров Михаил Николаевич, теперь на счете 3 292.15 ₽"], ["O", "B-WITHDRAW", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O"]),
    (["− 2 250.50 ₽ по номеру телефона +79955443322, Егорова Ольга Александровна, через СБП, теперь на счете 1 041.65 ₽"], ["O", "B-WITHDRAW", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O"]),
    (["− 1 500.75 ₽ списано с карты **** 3210, Никифоров Сергей Владимирович, теперь на счете -459.10 ₽"], ["O", "B-WITHDRAW", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O"]),
    (["− 2 300.00 ₽ по номеру телефона +79909876543, Мартынов Алексей Андреевич, через СБП, теперь на счете 1 500.10 ₽"], ["O", "B-WITHDRAW", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O"]),
    (["− 5 600.50 ₽ списано с карты **** 1234, Александрова Наталья Сергеевна, теперь на счете 12 100.50 ₽"], ["O", "B-WITHDRAW", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O"]),
    (["− 3 750.75 ₽ по номеру телефона +79913425768, Петров Петр Петрович, через СБП, теперь на счете 8 349.75 ₽"], ["O", "B-WITHDRAW", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O"]),
    (["− 1 100.00 ₽ списано с карты **** 8765, Михайлова Ольга Николаевна, теперь на счете 7 249.75 ₽"], ["O", "B-WITHDRAW", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O"]),
    (["− 4 200.50 ₽ по номеру телефона +79976543210, Захаров Дмитрий Александрович, через СБП, теперь на счете 3 049.25 ₽"], ["O", "B-WITHDRAW", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O"]),
    (["− 7 500.00 ₽ списано с карты **** 2345, Романов Сергей Николаевич, теперь на счете -4 450.75 ₽"], ["O", "B-WITHDRAW", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O"]),
    (["− 6 000.25 ₽ по номеру телефона +79987651234, Воробьев Антон Петрович, через СБП, теперь на счете -10 451.00 ₽"], ["O", "B-WITHDRAW", "B-AMOUNT", "I-AMOUNT", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O", "O"]),
    (["MIR-8894", "21:43", "перевод", "340р", "TINKOFF", "Баланс:", "234", "931.1р"], ["O", "O", "O", "B-WITHDRAW", "B-AMOUNT", "O", "O", "O", "O"]),
    (["MIR-8894", "21:43", "перевод", "5340р", "TINKOFF", "Баланс:", "23", "931.1р"], ["O", "O", "O", "B-WITHDRAW", "B-AMOUNT", "O", "O", "O", "O"]),
    (["MIR-8894", "21:45", "перевод", "2700р", "TINKOFF", "Баланс:", "22", "231.1р"], ["O", "O", "O", "B-WITHDRAW", "B-AMOUNT", "O", "O", "O", "O"]),
    (["MIR-8894", "21:45", "перевод", "2200р", "TINKOFF", "Баланс:", "11", "321.1р"], ["O", "O", "O", "B-WITHDRAW", "B-AMOUNT", "O", "O", "O", "O"]),
    (["MIR-8894", "11:45", "перевод", "2400р", "TINKOFF", "Баланс:", "12", "231.1р"], ["O", "O", "O", "B-WITHDRAW", "B-AMOUNT", "O", "O", "O", "O"]),
    (["MIR-8894", "22:45", "перевод", "2700р", "TINKOFF", "Баланс:", "21", "231.1р"], ["O", "O", "O", "B-WITHDRAW", "B-AMOUNT", "O", "O", "O", "O"]),
    (["MIR-8894", "22:45", "перевод", "2200р", "TINKOFF", "Баланс:", "11", "231.1р"], ["O", "O", "O", "B-WITHDRAW", "B-AMOUNT", "O", "O", "O", "O"]),
    (["MIR-8894", "21:45", "перевод", "2700р", "TINKOFF", "Баланс:", "21", "231.1р"], ["O", "O", "O", "B-WITHDRAW", "B-AMOUNT", "O", "O", "O", "O"]),
    (["MIR-8894", "21:45", "перевод", "2200р", "TINKOFF", "Баланс:", "11", "231.1р"], ["O", "O", "O", "B-WITHDRAW", "B-AMOUNT", "O", "O", "O", "O"]),
    (["MIR-8894", "11:45", "перевод", "2700р", "TINKOFF", "Баланс:", "12", "231.1р"], ["O", "O", "O", "B-WITHDRAW", "B-AMOUNT", "O", "O", "O", "O"]),
    (["MIR-8894", "11:45", "перевод", "2700р", "TINKOFF", "Баланс:", "11", "321.1р"], ["O", "O", "O", "B-WITHDRAW", "B-AMOUNT", "O", "O", "O", "O"]),
    (["MIR-8894", "11:45", "перевод", "2200р", "TINKOFF", "Баланс:", "11", "131.1р"], ["O", "O", "O", "B-WITHDRAW", "B-AMOUNT", "O", "O", "O", "O"]),
    (["MIR-8894", "11:45", "перевод", "2700р", "TINKOFF", "Баланс:", "12", "111.1р"], ["O", "O", "O", "B-WITHDRAW", "B-AMOUNT", "O", "O", "O", "O"]),

]


data = data_popolnenie + data_spisanie
sentences, labels = zip(*data)


tokenizer = Tokenizer(lower=False, oov_token="<OOV>")
tokenizer.fit_on_texts(sentences)
word_index = tokenizer.word_index

sequences = tokenizer.texts_to_sequences(sentences)
max_length = max(len(seq) for seq in sequences)
padded_sequences = pad_sequences(sequences, maxlen=max_length, padding='post')


label_encoder = LabelEncoder()
label_encoder.fit([label for sublist in labels for label in sublist])
encoded_labels = [label_encoder.transform(label) for label in labels]
padded_labels = pad_sequences(encoded_labels, maxlen=max_length, padding='post', value=-1)

num_labels = len(label_encoder.classes_)
padded_labels = np.array([tf.keras.utils.to_categorical(i, num_classes=num_labels) for i in padded_labels])


X_train, X_test, y_train, y_test = train_test_split(padded_sequences, padded_labels, test_size=0.2, random_state=42)


vocab_size = len(word_index) + 1

learning_rate = 5e-5
batch_size = 64
num_epochs = 50
dropout_rate = 0.6
lstm_units_1 = 128
lstm_units_2 = 64
regularization_rate = 1e-3


input_word_ids = tf.keras.layers.Input(shape=(max_length,), dtype=tf.int32, name="input_word_ids")
embedding_layer = tf.keras.layers.Embedding(input_dim=vocab_size, output_dim=128)(input_word_ids)

lstm_layer1 = tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(units=lstm_units_1, return_sequences=True,
                                                                kernel_regularizer=tf.keras.regularizers.l2(regularization_rate)))(embedding_layer)
batch_norm1 = tf.keras.layers.BatchNormalization()(lstm_layer1)
dropout_layer1 = tf.keras.layers.Dropout(dropout_rate)(batch_norm1)

lstm_layer2 = tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(units=lstm_units_2, return_sequences=True,
                                                                kernel_regularizer=tf.keras.regularizers.l2(regularization_rate)))(dropout_layer1)
batch_norm2 = tf.keras.layers.BatchNormalization()(lstm_layer2)
dropout_layer2 = tf.keras.layers.Dropout(dropout_rate)(batch_norm2)

dense_layer = tf.keras.layers.TimeDistributed(tf.keras.layers.Dense(num_labels, activation='softmax'))(dropout_layer2)

model = tf.keras.Model(inputs=input_word_ids, outputs=dense_layer)
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate),
              loss=tf.keras.losses.CategoricalCrossentropy(from_logits=False),
              metrics=['accuracy'])

model.summary()


tensorboard_callback = TensorBoard(log_dir="./logs")

model.fit(X_train, y_train, epochs=num_epochs, batch_size=batch_size, validation_data=(X_test, y_test), callbacks=[tensorboard_callback])
```
